```python
import pandas as pd

# 1. Ruta al archivo de referencia
path_csv = "../model_artifact/reference_data.csv"

try:
    # 2. Leer solo la primera fila para ir rápido
    df_temp = pd.read_csv(path_csv, nrows=1)
    
    print("="*60)
    print(f"COLUMNAS DETECTADAS EN: {path_csv}")
    print("="*60)
    
    # 3. Listar todas las columnas con su índice
    for i, col in enumerate(df_temp.columns):
        print(f" {i+1:3} | {col}")
        
    print("="*60)
    print(f"TOTAL: {len(df_temp.columns)} columnas.")
    
    # 4. Generar automáticamente el código para las 3 categorías
    print("\n>>> COPIA ESTO EN TU _schema_vers_entite (Ajusta los nombres si varían):")
    cols_interes = ['type_residence', 'objet_pret', 'type_pret']
    
    for c in cols_interes:
        if c in df_temp.columns:
            # Sacar valores únicos reales (leyendo toda la columna ahora)
            valores = pd.read_csv(path_csv, usecols=[c])[c].unique().tolist()
            print(f"{c:<25} = schema.{c},  # Valores en CSV: {valores}")
        else:
            print(f"# ⚠️ OJO: La columna '{c}' NO existe. Busca el nombre correcto en la lista de arriba.")

except FileNotFoundError:
    print(f"❌ Error: No se encuentra el archivo en {path_csv}")
```
